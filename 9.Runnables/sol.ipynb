```python
from abc import ABC, abstractmethod
```

```python
class Runnable(ABC):
    @abstractmethod
    def invoke(input_data):
        pass
```

```python
import random
class NakliLLM(Runnable):
    def __init__(self):
        print("LLM created")

    def invoke(self, prompt):

        response_list = [
            'Delhi is capital of India',
            'IPL is a cricket league',
            'AI stands for Artificial Intelligence'
        ]

        return{'respone': random.choice(response_list)}

    def predict(self, prompt):

        response_list = [
            'Delhi is capital of India',
            'IPL is a cricket league',
            'AI stands for Artificial Intelligence'
        ]

        return{'respone': random.choice(response_list)}
```

```python
class NakliPromptTemplate(Runnable):
    def __init__(self, template, input_variable):
        self.template = template
        self.input_variable = input_variable
    def invoke(self, input_dict):
        return self.template.format(**input_dict)

    def format(self, input_dict):
        return self.template.format(**input_dict)
```

### Chain Created , This is Standarised , Using abstraction of OOPs

```python
class NakliStrOutputParser(Runnable):
    def __init__(self):
        pass
    def invoke(self,input_data):
        return input_data['respone']
```

```python
class RunnableConnector(Runnable):

    def __init__(self, runnable_list):
        self.runnable_list = runnable_list

    def invoke(self, input_data):

        for runnable in self.runnable_list:
            input_data = runnable.invoke(input_data)

        return input_data
```

```python
template = NakliPromptTemplate(
    template = "Write a {length} poem about {topic}",
    input_variable = ['length','topic']
    
)
```

```python
parser = NakliStrOutputParser()
```

```python
llm = NakliLLM()
```

```python
# chain = RunnableConnector({template,llm})

chain = RunnableConnector([template,llm, parser])
```

### using { }, doesnot granteed that chain will be executed in same order

### therefore use [ ], this grantees execution in similar order

```python
chain.invoke({'length':'long','topic': 'india'})
```

## Creating multilevel Chain which uses multiple LLM 

```python
template1 = NakliPromptTemplate(
    template='Write a joke about {topic}',
    input_variable=['topic']
)
```

```python
template2 = NakliPromptTemplate(
    template = 'Explain the following joke {respone}',
    input_variable=['respone']

)
```

```python
llm = NakliLLM()
```

```python
parser = NakliStrOutputParser()
```

# Create 2 chain over here

```python
chain1 = RunnableConnector([template1,llm])
```

```python
chain2 = RunnableConnector([template2, llm,parser])
```

```python
final_chain = RunnableConnector([chain1,chain2])
```

```python
final_chain.invoke({'topic':'cricket'})
```
