```python
# Cell 1
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma

embedding = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python

from langchain_core.documents import Document
# from langchain_chroma import Chroma
# Create LangChain Documents for IPL playeres
doc1 = Document(
    page_content="Langchain helps developers build LLM application easily",
    # metadata={"team": "Royal Challengers Bangalore"}
)

doc2 = Document(
    page_content="Chroma is a Vector database optimized for LLM based seach",
    # metadata={"team": "Mumbai Indians"}
)

doc3 = Document(
    page_content="Embedding convert text into high-dimensional vectors",
    # metadata={"team": "Chennai Super Kings"}
)

doc4 = Document(
    page_content="Hugging face provides powerfull embedding models",
    # metadata={"team": "Mumbai Indians"}
)

doc5 = Document(
    page_content="Ravindra Jadeja is a dynamic all-rounder who contributes with both bat and ball. Representing Chennai Super Kings, his ...",
    # metadata={"team": "Chennai Super Kings"}
)
```

```python
docs = [doc1 ,doc2 , doc3, doc4, doc5]
```

```python
print(Chroma)
```

```python
vector_stores = Chroma.from_documents(
    documents= docs,
    embedding= embedding,
    collection_name="my_collection"
    
)
```

```python
vector_stores.get(include=['embeddings'])
```

### Convert vector store into a retriver


```python
retriever = vector_stores.as_retriever(search_kwargs={'k':1})
```

```python
query = "Who is virat kholi?"
result = retriever.invoke(query)
print(result)
```

```python
# Print retrived content
for i, doc in enumerate(result):
    print(f"\n ---Result {i+1} ---")
    print(f"Content: \n {doc.page_content}...") 
```
