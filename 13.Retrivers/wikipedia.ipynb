```python
from langchain_community.retrievers import WikipediaRetriever
```

```python
# Initializes the retrivers (optional : set languagge and top_k)
retriver = WikipediaRetriever(top_k_results=2, lang='en')
```

```python
# Define your query

query = " the  geopolitical history of india and pakistan from the independence "

# Get the relevant Wikipedia Documenst

docs = retriver.invoke(query)
```

```python
# Print retrived content
for i, doc in enumerate(docs):
    print(f"\n ---Result {i+1} ---")
    print(f"Content: \n {doc.page_content}...") 
```
