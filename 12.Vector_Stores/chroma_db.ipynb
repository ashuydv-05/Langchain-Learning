```python
from langchain_huggingface import HuggingFaceEmbeddings
embedding = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2")
from langchain_chroma import Chroma
```

```python

from langchain_core.documents import Document
# Create LangChain Documents for IPL playeres
doc1 = Document(
    page_content="Virat Kohli is one of the most successful and consistent batsmen in IPL history. Known for his aggressive batting style ...",
    metadata={"team": "Royal Challengers Bangalore"}
)

doc2 = Document(
    page_content="Rohit Sharma is the most successful captain in IPL history, leading Mumbai Indians to five titles. He's known for his calm ...",
    metadata={"team": "Mumbai Indians"}
)

doc3 = Document(
    page_content="MS Dhoni, famously known as Captain Cool, has led Chennai Super Kings to multiple IPL titles. His finishing skills, wicket ...",
    metadata={"team": "Chennai Super Kings"}
)

doc4 = Document(
    page_content="Jasprit Bumrah is considered one of the best fast bowlers in T20 cricket. Playing for Mumbai Indians, he is known for his ...",
    metadata={"team": "Mumbai Indians"}
)

doc5 = Document(
    page_content="Ravindra Jadeja is a dynamic all-rounder who contributes with both bat and ball. Representing Chennai Super Kings, his ...",
    metadata={"team": "Chennai Super Kings"}
)
```

```python
docs = [doc1 ,doc2 , doc3, doc4, doc5]
```

```python
vector_store = Chroma(
    # after entering persist_dicretory , automatically folder gets created.
    embedding_function= embedding,
    persist_directory='my_chroma_db', 
    collection_name='sample'
)
```

### ADD DOCUMENTS

#### vector_store.add_documents(docs) :--> embedding is done and IDs are created shown below

```python
vector_store.add_documents(docs)
```

###     VIEW DOCUMENTS


```python
vector_store.get()
```

```python
vector_store.get(include = [ 'documents','metadatas','embeddings'])
```

```python

vector_store.get(include=["embeddings"])
```

### Search Documents

```python
vector_store.similarity_search(
    query="Who among these are a bowler?",
    k = 2
)

# k is the parameter --> kitne similar result output me chahiye
# k =1 , 1 result in OP , k=2 2 result in OP
```

```python
vector_store.similarity_search_with_score(
    query="Who among these are a bowler?",
    k = 2
)

# score represent the distance, jitna kam utna accha
```

### Meta-Data Filtering

```python
vector_store.similarity_search_with_score(
    query="",
    filter={'team':"Chennai Super Kings"}
)
```

### UPDATE DOCUMENTS

```python
updated_doc1 = Document(
    page_content= "Virat kholi, the former captaon of Royal Challenger Bangalore RCB, is renowed " \
    "for his aggressive leadership and consistency play",
    metadata = {'team':"Royal Challengers Bangalore"}
)

vector_store.update_document(document_id="3d9af13d-d889-4c2a-8779-d3bdb9dcce1f", document=updated_doc1)
```
